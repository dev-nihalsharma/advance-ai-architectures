```python
import torch
import torch.nn as nn
```

```python
VGG_NET_ARC = {
    'VGG16': [64,64,'M',128,128,'M',256,256,256,'M',512,512,512,'M',512,512,512,'M']
}
```

```python
class VGG_NET(nn.Module):
    def __init__(self, num_classes,in_channels=3):
        super(VGG_NET,self).__init__()
        self.in_channels = in_channels
        self.conv_layers = self.create_conv_layers(VGG_NET_ARC['VGG16'])
        self.fcs = nn.Sequential(
         nn.Linear(512*7*7,4096),
         nn.ReLU(),
         nn.Dropout(p=0.5),
         nn.Linear(4096,4096),
         nn.ReLU(),
         nn.Dropout(p=0.5),
         nn.Linear(4096, num_classes)
        )
        
    def forward(self,x):
        x = self.conv_layers(x)
        print(x.shape)
        x = x.reshape(x.shape[0], -1)
        x = self.fcs(x)
        
        return x
        
    def create_conv_layers(self, architechture):
        in_channels = self.in_channels 
        layers = []
        
        for x in architechture:
            if type(x) == int:
                layers += [
                    nn.Conv2d(in_channels=in_channels, out_channels=x,kernel_size=(3,3), stride=(1,1), padding=(1,1)),
                    nn.BatchNorm2d(x),
                    nn.ReLU()
                ]
                
                in_channels = x
                
            if x == "M":
                layers += [nn.MaxPool2d(kernel_size=(2,2), stride=(2,2))]
                
        print(len(layers))
        
        return nn.Sequential(*layers)
                
```

```python
model = VGG_NET(in_channels=3, num_classes=1000)
x = torch.rand(1,3,224,224)
print(model(x).shape)
```
